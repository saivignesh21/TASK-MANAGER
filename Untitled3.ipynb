```python
# STEP 2: Unzip and import libraries
import zipfile, os, glob
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, measure, filters, morphology
import pandas as pd

zip_path = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("frames")

frame_paths = sorted(glob.glob("frames/*.tif"))
print("Total frames found:", len(frame_paths))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import pandas as pd

diameters, heights, frames_idx = [], [], []

for i, path in enumerate(frame_paths):
    frame = io.imread(path)
    frame = cv2.normalize(frame, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Blur to reduce noise
    blur = cv2.GaussianBlur(frame, (5,5), 0)

    # Edge detection
    edges = cv2.Canny(blur, 30, 150)

    # Find contours
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) > 0:
        # Pick the largest contour by area (the droplet)
        c = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(c)
        diameters.append(w)
        heights.append(h)
        frames_idx.append(i)
    else:
        diameters.append(0)
        heights.append(0)
        frames_idx.append(i)

    # Debug display every 50 frames
    if i % 50 == 0:
        vis = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
        cv2.drawContours(vis, [c], -1, (0,255,0), 1)
        plt.figure(figsize=(8,4))
        plt.subplot(1,2,1)
        plt.imshow(frame, cmap='gray')
        plt.title(f'Frame {i}')
        plt.subplot(1,2,2)
        plt.imshow(vis)
        plt.title('Detected Contour')
        plt.show()

print("Contour-based analysis complete!")
```

```python
plt.figure(figsize=(10,5))
plt.plot(frames_idx, diameters, label='Droplet Diameter (pixels)', lw=2)
plt.plot(frames_idx, heights, label='Droplet Height (pixels)', lw=2)
plt.xlabel("Frame Number")
plt.ylabel("Pixels")
plt.title("Droplet Diameter & Height Variation (Contour Detection)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# STEP 5: Save results to CSV
df = pd.DataFrame({
    "Frame": frames_idx,
    "Diameter (px)": diameters,
    "Height (px)": heights
})
df.to_csv("droplet_analysis.csv", index=False)
from google.colab import files
files.download("droplet_analysis.csv")
```
